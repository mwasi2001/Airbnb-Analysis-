# **Finding the Outliars & Remove It Potentially**

```python
sns.boxplot(df_airbnb['price'])
```

```python
# outliear foinding
import numpy as np
q1 = np.percentile(df_airbnb['price'],25)
q3 = np.percentile(df_airbnb['price'],75)

iqr = q3 - q1

lower_bound = iqr - 1.5*iqr
upper_bound = iqr + 1.5*iqr
df_airbnb[(df_airbnb['price']< lower_bound)| (df_airbnb['price']> upper_bound)]['price'].unique()
```

```python
# not outliear (remove)
import numpy as np
q1 = np.percentile(df_airbnb['price'],25)
q3 = np.percentile(df_airbnb['price'],75)

iqr = q3 - q1

lower_bound = iqr - 1.5*iqr
upper_bound = iqr + 1.5*iqr
data_free_outliear =df_airbnb[(df_airbnb['price']> lower_bound)| (df_airbnb['price']< upper_bound)]
data_free_outliear
```
